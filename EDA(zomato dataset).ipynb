```python
#reading the excel file
df_country=pd.read_excel("Country-Code.xlsx")
df_country.head()                       
```

```python
df.columns
```

```python
#joining the csv file with Excel file(Country Code)
final_df=pd.merge(df,df_country,on="Country Code",how='left')
```

```python
final_df.head(5)
```

```python
final_df.dtypes
```

```python
final_df.columns
```

```python
#how many records are there with respective countries
final_df.Country.value_counts()
```

```python
country_name=final_df.Country.value_counts().index
```

```python
country_name
```

```python
country_values=final_df.Country.value_counts().values
```

```python
# want to check top 3 countries that uses zomato
##pie chart
plt.pie(country_values[:3],labels=country_name[:3],autopct="%1.2f%%")
```

```python
#Conclusion: It is clear from the pie chart that India has maximum transaction.
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
ratings
```

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12, 6)
sns.barplot(x="Aggregate rating",y="Rating count",data=ratings)
```

```python
sns.barplot(x="Aggregate rating",y="Rating count",hue='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','green'])
```

Observation:

1.Not Rated count is very high
2.Maximum number of rating are between 2.5 to 3.4

```python
## Count plot
sns.countplot(x="Rating color",data=ratings,palette=['blue','red','orange','yellow','green','green'])
```
